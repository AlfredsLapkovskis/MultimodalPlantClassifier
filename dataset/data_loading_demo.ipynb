```python
import sys
import os

root = os.path.abspath("..")
sys.path.append(root)

from config import Config
from common.constants import MODALITY_LIST
import matplotlib.pyplot as plt
```

```python
config = Config(os.path.join(root, "config.json"))
```

```python
from loading import load_unimodal_dataset

ds = load_unimodal_dataset(
    config,
    "Fruit",
    "train",
    shuffle=True,
    batch_size=16,
)

for batch in ds.take(1):
    plt.imshow(batch[0][0])
```

```python
from loading import load_multimodal_dataset

mm_ds = load_multimodal_dataset(
    config,
    "validation",
    modalities=MODALITY_LIST,
    shuffle=True,
    dropout=0.125,
    batch_size=16,
)

for batch in mm_ds.take(1):
    images = batch[0]
    for i, img in enumerate(images.values()):
        plt.subplot(1, 4, i + 1)
        plt.imshow(img[0])
```
